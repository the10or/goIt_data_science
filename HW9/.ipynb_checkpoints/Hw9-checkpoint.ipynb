```python
import tensorflow as tf
from tensorflow import keras
```

## 1.Підготовка та аналіз даних

### Завантажуємо датасет та поділяємо його на тренувальний та тестовий набори

```python
from keras.datasets import fashion_mnist
dataset = fashion_mnist.load_data()
(x_train, y_train), (x_test, y_test) = dataset
print(f"x_train shape: {x_train.shape}, y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}, y_test shape: {y_test.shape}")
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

### Виведемо випадкове зображення з тренувального датасету

```python
import random
import matplotlib.pyplot as plt
class_number = random.randint(0, len(x_train))
plt.imshow(x_train[class_number], cmap='binary')
plt.title(class_names[y_train[class_number]])
plt.axis('off')
plt.colorbar()
plt.gca().grid(False)
```

### Нормалізуємо дані

```python

x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

### Виведемо перші 25 зображень

```python
plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(5, 5, i+1)
  plt.xticks([])
  plt.yticks([])
  # plt.grid('off')
  plt.imshow(x_train[i], cmap="binary")
  plt.title(class_names[y_train[i]])
plt.show()
```

## 2. Побудова та вибір моделей

### Будуємо першу примітивну модель з одним вхідним, одним прихованим та одним вихідним шарами

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
def plot_history(history):
  history_dict = history.history
  loss_values = history_dict["loss"]
  val_loss_values = history_dict["val_loss"]


  acc_values = history_dict["accuracy"]
  val_acc_values = history_dict["val_accuracy"]


  epochs = range(1, len(acc_values) + 1)
  plt.figure(figsize=(8, 5))
  plt.plot(epochs, loss_values,  color="r",  label="Training loss")
  plt.plot(epochs, acc_values, "b", label="Training accuracy")
  plt.plot(epochs, val_loss_values, "r:", label="Validation loss")
  plt.plot(epochs, val_acc_values, "b:", label="Validation accuracy")

  plt.title("Training loss and accuracy")
  plt.xlabel("Epochs")
  plt.ylabel("Loss")
  plt.legend()
  plt.show()

plot_history(history)
```

Судячи з графіку кривих навчання, модель навчається не дуже ефективно. Хоча на навчальних даних accuracy та loss відповідно зростають, та спадають, відповідні показники на валідаційних даних або суттєво не змінюються, або навіть погіршуються.

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('\nTest accuracy:', test_acc)
```

## Передбачення для тестового набору даних і його візуалізація

```python
predictions = model.predict(x_test)
```

## Визначаємо функцію для візуалізації результатів класифікації

```python
import numpy as np
def visualize_prediction(predictions=predictions):
  plt.figure(figsize=(10,10))
  for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('on')
    label =  np.random.randint(0, len(predictions))
    plt.imshow(x_test[label], cmap="binary")


    predicted_label = np.argmax(predictions[label])
    true_label = y_test[label]
    if predicted_label == true_label:
      color = "green"
    else:
      color = "red"
    plt.xlabel("{}({})".format(class_names[predicted_label],
                                        class_names[true_label]),
                                        color=color)

visualize_prediction(predictions)
```

## Розширена модель з додатковими шарами

```python
model_adv = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),

    # keras.layers.Dense(256, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
    ])

model_adv.summary()
```

```python
optimizer_adam = keras.optimizers.Adam(learning_rate=0.001)
model_adv.compile(optimizer=optimizer_adam,
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

```

```python
history_adv = model_adv.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
plot_history(history_adv)
```

```python
test_loss, test_acc = model_adv.evaluate(x_test, y_test)
print('\nTest accuracy:', test_acc)
```

з графіків видно, що модель менше схильна до перенавчання, а показники точності і втрат показують схожі тенденції на навчальних і валідаційних даних, або навіть майже збігаються. Також вони близькі до результатів, отриманих на тестових даних.

```python
predictions_adv = model_adv.predict(x_test)
visualize_prediction(predictions_adv)
```

## Знаходження оптимальних параметрів моделі за допомогою callback-функцій

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
early_stopping = EarlyStopping(patience=10, monitor="val_loss")
checkpoint = ModelCheckpoint("best_model.keras", monitor="val_loss", save_best_only=True, verbose=1)
callbacks = [
  early_stopping,
  checkpoint
]

history_with_callbacks = model_adv.fit(x_train, y_train, validation_split=0.2, epochs=30, callbacks=callbacks, batch_size=32)
```

```python
plot_history(history_with_callbacks)
```

З графіку видно, що після 5 епохи результати на валідаційному наборі перестали показувати тенденцію до покращення, якщо орієнтуватись на validation loss. Тому через 10 епох тренування було зупинено завдяки коллбеку early stopping, а параметри моделі 5 епоху було збережено як best model.

```python
best_model = keras.models.load_model("best_model.keras")
best_model.summary()
```

```python
best_model_loss, best_model_acc = best_model.evaluate(x_test, y_test)
print('\nTest accuracy:', best_model_acc)
```

```python
visualize_prediction(best_model.predict(x_test))
```
