## Завантаження та підготовка датасету

```python
import matplotlib.pyplot as plt
from keras.datasets import imdb
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense, LSTM, Bidirectional
```

```python
max_features = 10000 
maxlen = 500  
batch_size = 32
epochs_num = 10
```

```python
start_char = 1
oov_char = 2  
index_from = 3
```

```python
(input_train, y_train), (input_test, y_test) = imdb.load_data(num_words=max_features, maxlen=maxlen, start_char=start_char, oov_char=oov_char, index_from=index_from)
i
```

```python
input_train[0]
```

```python
len(input_train)
```

```python
# vocabulary
word_index = imdb.get_word_index()
word_index
```

```python
# Reverse the word index to obtain a dict mapping indices to words
# And add `index_from` to indices to sync with `x_train`
inverted_word_index = dict(
    (i + index_from, word) for (word, i) in word_index.items()
)
# Update `inverted_word_index` to include `start_char` and `oov_char`
inverted_word_index[start_char] = "[START]"
inverted_word_index[oov_char] = "[OOV]"
# Decode the first sequence in the dataset
decoded_sequence = " ".join(inverted_word_index[i] for i in input_train[0])
decoded_sequence
```

```python
# padding
input_train = pad_sequences(input_train, maxlen=maxlen)
input_test = pad_sequences(input_test, maxlen=maxlen)
# truncation - зрізання речення за межами максимальної довжини
# Застосування заздалегідь обраної максимальної довжини до послідовних даних тренувального та тестового наборів
```

```python
input_train[0]
```

```python
len(input_train[0])
```

## SimpleRNN модель

```python
model = Sequential(name='Simple RNN')

model.add(Embedding(max_features, 32))
model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))

model.name
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

```python
model.summary()
```

```python
history = model.fit(input_train, y_train, epochs=epochs_num, batch_size=128, validation_split=0.3)
# Запускаємо навчання моделі на тренувальних даних для 5 епох з розміром пакету 128 та розбиттям на тренувальну
```

## Функція візуалізації результатів

```python
def plot_history(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs = range(1, len(acc) + 1)
    
    plt.plot(epochs, acc, 'r--',  label='Точність на тренуванні')  # Графік точності на тренувальних данних
    plt.plot(epochs, val_acc, 'b', label='Точність на валідації')  # Графік точності на валідаційних данних
    plt.title(f'Точність тренування і валідації моделі {history.model.name}')  # Заголовок графіку
    plt.legend()  # Показати легенду
    
    plt.figure()
    
    plt.plot(epochs, loss, 'r--', label='Втрати на тренуванні')  # Графік втрат на тренувальних данних
    plt.plot(epochs, val_loss, 'b', label='Втрати на валідації')  # Графік втрат на валідаційних данних
    plt.title(f'Втрати тренування і валідації моделі {history.model.name}')  # Заголовок графіку
    plt.legend()  # Показати легенду
    
    plt.show() 
```

```python

plot_history(history)# Відобразити графіки
```

```python
model.evaluate(input_test, y_test)
```

```python
predicted = model.predict(input_test[0:10])
for p, l in zip(predicted, y_test[0:10]):
    print(round(p[0]) , l, round(p[0]) == l)
    
```

```python
lstm_model = Sequential(name="LSTM")
lstm_model.add(Embedding(max_features, 32))
lstm_model.add(LSTM(32))
lstm_model.add(Dense(1, activation='sigmoid'))
lstm_model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
lstm_model.summary()
```

```python
lstm_history = lstm_model.fit(input_train, y_train, epochs=epochs_num, batch_size=128, validation_split=0.3)
```

```python
plot_history(lstm_history)
```

```python
lstm_model.evaluate(input_test, y_test)
```

```python
lstm_predicted = lstm_model.predict(input_test[0:10])
for p, l in zip(lstm_predicted, y_test[0:10]):
    print(round(p[0]) , l, round(p[0]) == l)
```

```python
lstm_bidirectional_model = Sequential()
lstm_bidirectional_model.add(Embedding(max_features, 32))
lstm_bidirectional_model.add(Bidirectional(LSTM(32)))
lstm_bidirectional_model.add(Dense(1, activation='sigmoid'))
lstm_bidirectional_model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
lstm_bidirectional_model.summary()
```

```python
lstm_bidirectional_history = lstm_bidirectional_model.fit(input_train, y_train, epochs=epochs_num, batch_size=128, validation_split=0.3)
```

```python
plot_history(lstm_bidirectional_history)
```

```python
lstm_bidirectional_model.evaluate(input_test, y_test)
```

```python
lstm_bidirectional_predicted = lstm_bidirectional_model.predict(input_test[0:10])
for p, l in zip(lstm_bidirectional_predicted, y_test[0:10]):
    print(round(p[0]) , l, round(p[0]) == l)
```
