# Частина перша: Знайомство з Pandas.


```python

import numpy as np
import pandas as pd
```

## Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
df = pd.read_html(url, match="Коефіцієнт народжуваності в регіонах України", header=0, thousands=".", decimal=",")[0]
```

## Вивести перші рядки таблиці за допомогою методу head

```python
df.head()
```

## Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df.shape
```

## Замініть у таблиці значення "—" на значення NaN

```python
df.replace('—', np.nan, inplace=True)
df
```

## Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

## Замініть типи нечислових колонок на числові.

```python
df = df.astype({"2014": "float", "2019": "float"})
df.dtypes
```

## Порахуйте, яка частка пропусків міститься в кожній колонці 

```python
df.isna().sum() / len(df)
```

## Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df.drop(df.tail(1).index, inplace=True)
df
```

## Замініть відсутні дані в стовпцях середніми значеннями цих стовпців крім стовпця "Регіон"

```python
for column in df.columns[1:]:
    df.fillna({column: df[column].mean()}, inplace=True)
df
```

## Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
high_fertility_list_2019_df = df[df["2019"] > df["2019"].mean()][["Регіон", "2019"]]
high_fertility_list_2019_df
```

## У якому регіоні була найвища народжуваність у 2014 році?

```python
highest_fertility_region_2014 = df[df["2014"] == df["2014"].max()][["Регіон"]]
highest_fertility_region_2014
```

## Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt
blue = "DodgerBlue"
plt.figure(figsize=(20, 10))
plt.title("Народжуваність по регіонам в 2019 році", fontsize=40, fontweight="bold", pad=20, loc="center", color=blue)
plt.xlabel("Регіон", fontsize=20, fontweight="bold", labelpad=20, color=blue)
plt.xticks(rotation=65, color=blue)
plt.ylabel("Народжуваність", fontsize=20, fontweight="bold", labelpad=20, color=blue,  )
plt.bar(df["Регіон"], df["2019"], color=blue, edgecolor="gold", hatch="*//////")
plt.grid(True)
plt.savefig("Hw2.1.png")
plt.show()

```

```python
plt.figure(figsize=(20, 10))
plt.title("Народжуваність по регіонам в 2019 році", fontsize=30, color="maroon", fontweight="bold", pad=20, loc="center")
plt.pie(df["2019"], labels=df["Регіон"], autopct='%1.1f%%', wedgeprops={'edgecolor': 'darkslategray'},)
plt.show()
```

```python
plt.figure(figsize=(30, 15))
plt.grid(True)
plt.title("Народжуваність по регіонам в 1950 році", fontsize=40, fontweight="bold", pad=20, loc="center",
          color="maroon")
plt.scatter(df["Регіон"], df["1950"], color="mediumorchid", edgecolor="gold", s=200)
plt.xlabel("Регіон", fontsize=20, fontweight="bold", labelpad=20, color="maroon")
plt.ylabel("Народжуваність", fontsize=20, fontweight="bold", labelpad=20, color="maroon")
plt.xticks(rotation=35, ha="right", color="maroon", fontsize=16)
plt.yticks(color="maroon", fontsize=16)
plt.show()
```
