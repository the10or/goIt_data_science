## Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
cols_to_drop = []
for col in df.columns:
    if col != "Мова програмування" and df[col].isnull().any():
        cols_to_drop.append(col)

df = df.drop(columns=cols_to_drop)

df
```

## Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

```python
df.isnull().sum() / len(df)
```

```python
df = df.dropna()
df
```

## Створіть нову таблицю *python_data*, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування *Python*

```python
python_data = df[df["Мова програмування"] == "Python"]
python_data
```

## Визначте розмір таблиці *python_data* за допомогою методу *shape*

```python
python_data.shape
```

## Використовуючи метод *groupby*, виконайте групування за стовпчиком "Посада"

```python
python_data_grouped = python_data.groupby("Посада")
python_data_grouped.groups.keys()
```

## виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата на місяць"

```python
grouped_df = python_data_grouped.agg({"Зарплата на місяць": ["min", "max"]})
grouped_df
```

## Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(group):
    return pd.Series({"avg": group["Зарплата на місяць"].mean()}).round(2)


grouped_df = pd.concat([grouped_df, python_data_grouped.apply(fill_avg_salary, include_groups=False)], axis=1)

grouped_df

```

## Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
grouped_df["avg"].describe().round(2)
```

```python
grouped_df.to_csv("python_data.csv", encoding="utf-8", sep=",")
```

## Візуалізуалізації

```python


# Додамо стовпці для максимального та мінімального значень до вашого датафрейму
grouped_df['max'] = grouped_df.max(axis=1)
grouped_df['min'] = grouped_df.min(axis=1)

fig, ax = plt.subplots(figsize=(10, 5))
ax.set_facecolor("cornsilk")
ax.grid(color="gray", alpha=0.5, linestyle="--")

# Відобразимо дані avg, max та min для кожної посади
ax.stem(grouped_df.index, grouped_df['avg'], markerfmt="^", linefmt="chocolate", label='avg')
ax.stem(grouped_df.index, grouped_df['max'], markerfmt="o", linefmt="green", label='max')
ax.stem(grouped_df.index, grouped_df['min'], markerfmt="s", linefmt="red", label='min')

ax.set_xlabel("Посада", fontsize=12, fontweight="bold", color="chocolate", labelpad=10)
ax.set_ylabel("Зарплата на місяць", fontsize=12, fontweight="bold", color="chocolate", labelpad=10)
ax.set_title("Зарплата спеціалістів по посадах", fontsize=14, fontweight="bold", color="chocolate", pad=20)
ax.tick_params(axis='x', rotation=35, labelcolor='chocolate')
ax.tick_params(axis='y', labelcolor='chocolate')
ax.legend()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

sns.set_style("darkgrid")

colors = sns.color_palette("husl", 3)

plt.figure(figsize=(10, 6))
sns.lineplot(data=grouped_df[['max', 'avg', 'min']], markers=True, palette=colors)

plt.title("Зарплата спеціалістів по посадах max, avg, min", color=colors[1], fontsize=20, fontweight='bold', pad=20)
plt.xticks(rotation=35, ha="right", color=colors[1], fontsize=12)
plt.xlabel("Посада", color=colors[1], fontsize=12, fontweight='bold', labelpad=10)
plt.ylabel("Зарплата на місяць", color=colors[1], fontsize=12, fontweight='bold', labelpad=10)

plt.show()
```
